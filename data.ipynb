```python
import os
import pandas as pd

folder_path = r"C:\Users\pulib\Desktop\clgcutoff"
dance_df = pd.DataFrame()

for file in os.listdir(folder_path):
    if file.endswith(".html"):
        file_path = os.path.join(folder_path, file)
        try:
            df = pd.read_html(file_path)[0]
            # Use the file name (without extension) as the source
            df['Source'] = os.path.splitext(file)[0]
            dance_df = pd.concat([dance_df, df], ignore_index=True)
            print(f"✅ Added: {file}")
        except Exception as e:
            print(f" Error reading {file}: {e}")

    # Reset index as S.NO
if not dance_df.empty:
    dance_df.index += 1
    dance_df.rename_axis("index.NO", inplace=True)

# Show   first few rows
print("\nFinal Combined DataFrame:")
dance_df.tail()
```
